# Análisis sobre el PBI de los países miembros plenos del Mercosur (Período 1991 - 2020)

En el presente estudio busco visualizar la situación actual en cuanto al PBI de los países miembros del Mercosur luego del impacto por la pandemia de Covid-19 y por otro lado analizar la evolución de este indicador desde la fecha de creación de este bloque comercial (1991).

## Importo las líbrerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set()
```

## PBI año 2020 de los países miembros plenos del Mercosur

```python
df = pd.read_excel('PBI_paises.xlsx')
df.head()
```

```python
filtro1 = df[df['Country Name'].isin(['Argentina','Brasil','Uruguay','Paraguay'])]
filtro1.head()
```

```python
filtro2 = filtro1[['Country Name', '2020']]
filtro2
```

```python
PBI = filtro2.sort_values('2020', ascending=True)
PBI
```

```python
# Gráfico de barras con el detalle del PBI de cada una de los países al año 2020
```

```python
plt.figure(figsize=(10,5), dpi=90)
sns.barplot(PBI['Country Name'],PBI['2020'])
plt.xlabel('Países') 
plt.ylabel('PBI (US$)')  
plt.title('PBI de los países miembros del Mercosur año 2020', fontsize=20)
pass
```

## PBI per capita en los países miembros del mercosur año 2020

```python
df = pd.read_excel('PBI_percapita.xlsx')
df.head()
```

```python
filtrado = df[df['Country Name'].isin(['Argentina','Brasil','Uruguay','Paraguay'])]
filtrado.head()
```

```python
filtrado2 = filtrado[['Country Name', '2020']]
filtrado2
```

```python
percapita = filtrado2.sort_values('2020', ascending=True)
percapita
```

```python
plt.figure(figsize=(10,5), dpi=90)
sns.barplot(percapita['Country Name'],percapita['2020'])
plt.xlabel('Países') 
plt.ylabel('PBI per capita (US$)')  
plt.title('PBI per capita de los países miembros del Mercosur año 2020', fontsize=20)
pass
```

## Evolución del PBI per capita de los países miembros del mercosur período 1991 - 2020

```python
depuracion = filtrado[['Country Name', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999',
                  '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
                  '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']]
depuracion.head()
```

```python
depuracion.to_excel("analisis.xlsx") 
```

```python
avance = pd.read_excel('analisis.xlsx')
avance.head()
```

```python
plt.figure(figsize=(10,4), dpi=90)

# País 1 
x1 = avance['Año']
y1 = avance['Argentina']

# País 2
x2 = avance['Año']
y2 = avance['Brasil']

# País 3 
x3 = avance['Año']
y3 = avance['Paraguay']

# País 4 
x4 = avance['Año']
y4 = avance['Uruguay']

plt.plot(x1,y1, marker='o', linestyle='-.', color='c', label='Argentina')
plt.plot(x2,y2, marker='o', linestyle='-.', color='g', label='Brasil')
plt.plot(x3,y3, marker='o', linestyle='-.', color='r', label='Paraguay')
plt.plot(x4,y4, marker='o', linestyle='-.', color='b', label='Uruguay')
plt.title('Evolución del PBI per capita en el período 1991-2020', fontsize=20)
plt.ylabel('PBI per capita (US$)')
plt.legend(loc='upper left')
plt.show()
```
