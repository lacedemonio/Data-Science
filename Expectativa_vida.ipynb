## Análisis sobre la evolución de los niveles de expectativa de vida de la población luego de la Revolución Industrial (Período 1800-2016)

En el presente estudio buscare analizar la evolución del indicador de expectativa de vida de la población en ciertos países desarrollados de Europa en primer término y luego transpolar el análisis a lo sucedido con un grupo específico de países de América. La idea de dicho estudio es visualizar la progresión de este indicador social en estos dos grupos de países y cuanto fue afectado por los grandes acontecimientos que marcaron el siglo XIX y XX (Revolución Industrial, guerras napoleónicas, descolonización de América, Guerras Mundiales, Guerras Civiles, Crisis del '29, etc.).

### Importo las librerías con las cuales voy a trabajar

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set()
from IPython.display import Image
```

### Lectura de la base de datos e información de la misma

```python
df = pd.read_csv('Life expectancy.csv')
df.head()
```

### Selecciono mi primer grupo de países a analizar (Naciones desarrolladas de Europa)

```python
Europa = df[df['Entity'].isin(['Germany', 'France', 'United Kingdom', 'Italy', 'Spain', 'Switzerland'])]
Europa.head()
```

```python
Alemania = df[df['Entity'] == 'Germany']
Francia = df[df['Entity'] == 'France']
UK = df[df['Entity'] == 'United Kingdom']
Italia = df[df['Entity'] == 'Italy']
Spain = df[df['Entity'] == 'Spain']
Suiza = df[df['Entity'] == 'Switzerland']
```

```python
plt.figure(figsize=(12,4), dpi=90)

# País 1 
x1 = Alemania['Year']
y1 = Alemania['Life expectancy']

# País 2
x2 = Francia['Year']
y2 = Francia['Life expectancy']

# País 3 
x3 = UK['Year']
y3 = UK['Life expectancy']

# País 4 
x4 = Italia['Year']
y4 = Italia['Life expectancy']

#País 5
x5 = Spain['Year']
y5 = Spain['Life expectancy']

# País 6 
x6 = Suiza['Year']
y6 = Suiza['Life expectancy']

plt.plot(x1,y1, marker='o', linestyle='-.', color='black', label='Alemania')
plt.plot(x2,y2, marker='o', linestyle='-.', color='blue', label='Francia')
plt.plot(x3,y3, marker='o', linestyle='-.', color='orange', label='Reino Unido')
plt.plot(x4,y4, marker='o', linestyle='-.', color='green', label='Italia')
plt.plot(x5,y5, marker='o', linestyle='-.', color='yellow', label='España')
plt.plot(x6,y6, marker='o', linestyle='-.', color='red', label='Suiza')
plt.title('Evolución de la expectativa de vida en el período 1800-2016', fontsize=20)
plt.ylabel('Años de expectativa de vida')
plt.legend(loc='upper left')
plt.show()
```

### Conclusiones de este primer grupo

Tal como podemos apreciar en los gráficos la expectativa de vida de la población de este conjunto de países de Europa comienza a crecer de manera rápida y sostenida en el último cuarto del siglo XIX y solo se ve interrumpida por el advenimiento de los grandes acentecimientos que marcaron el siglo XIX y XX, tales como la "primera guerra mundial", la "crisis del '29" y la "segunda guerra mundial" por mencionar los más importantes. En algunos países el impacto de estos sucesos fue más explosivo que en otros ya que no a todos los afecto por igual, tal como se nota en el caso de Alemania (fue el que mayor perdidas humanas y materiales sufrió en la segunda guerra mundial) o el caso de Suiza (se mantuvo neutral durante esta contienda).

El caso de España fue influido por otro acontecimiento que impacto de pleno en la calidad y expectativa de vida en los años ´30 del siglo XX que fue la cruenta guerra civil que debió atravesar.

Francia es interesante ver las fluctuaciones que presenta también en el siglo XIX, debido a la gran cantidad de guerras y revueltas en las que estuvo envuelta como nación (guerras napoleónicas, guerra Franco-Prusiana, Comuna de París, entre otras), lo que le impidio mantener este indicador de una manera más estable como las demás naciones de Europa estudiadas.

Pese a estos eventos enunciados la tendencia siempre fue al alza en todas las naciones y al final de la línea, al llegar al último año de análisis (2016) se puede apreciar como todos los países se mantienen en valores similares muy por encima de los que se manejaban tan solo un siglo atrás, muestra cabal de como la tecnología y la evolución de la ciencia moderna ha permitido aumentar las expectativas de vida en estos países desarrollados de Europa.
